```python
import requests
from bs4 import BeautifulSoup as bsp
import pandas as pd
url="https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
page=requests.get(url)
soup=bsp(page.content,'lxml')
soup
```

```python
table_data=soup.find_all(class_="wikitable sortable tpl-blanktable plainrowheaders")
len(table_data)
```

```python
covid_data=table_data[0]
covid_data
```

```python
head=[h.get_text().strip().split('\n\n') for h in covid_data.find_all('tr')]
head
```

```python
len(head)
```

```python
head[0]
```

```python
head[1]
```

```python
region=[n[0] for n in head[1:]]
region
```

```python
d={}
for i in head[0]:
    d[i]=[]
d
```

```python
len(head[0])
```

```python
for h in head[1:]:
    print(h)
```

```python
for h in head[1:]:
    field=h[1].split()
#     print(field)
    
    d['Region[27]'].append(h[0])
    d['Total cases'].append(field[0])
    d['Total deaths'].append(field[1])
    d['Cases per million'].append(field[2])
    d['Deaths per million'].append(field[3])
    d['Current weekly cases'].append(field[4])
    d['Current weekly deaths'].append(field[5])
    d['Population millions'].append(field[6])
    d['Vaccinated\xa0%[28]'].append(field[7])
d
```
